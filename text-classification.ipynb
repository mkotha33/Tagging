```python
from bs4 import BeautifulSoup
import urllib.request
#website to get data from
response = urllib.request.urlopen('https://www.math-only-math.com/word-problems-on-measuring-length.html')
html = response.read()
#print(html)
```

```python
#12.2.14 type dates
def checkone(x):
    count = 0
    c1 =0
    if x.count('.')==2:
        for i in x:
            if i.isdigit():
                count+=1
        if(count>=6):
            for i in range(len(x)):
                if x[i]=='.' and x[i+1].isdigit():
                    if (i-1>=0):
                        if x[i-1].isdigit():
                           c1 +=1
            if c1>0:
                return 1
            else:
                return 0
        else:
            return 0
    else:
        return 0
#day and year check
def check1(x):
    p = 0
    for i in x:
        if i.isdigit():
            p+=1
    if p==2 or p==1:
        return 1
    else:
        return 0
def check2(x):
    p = 0
    for i in x:
        if i.isdigit():
            p+=1
    if p==4:
        return 1
    else:
        return 0
def check3(x):
    l=0
    t=0
    for i in x:
        if i.isdigit():
            l+=1
    for i in x:
        if i.isalpha():
            t+=1
    c=0
    if l==4 and t==0:
        for i in x:
            if c<4:
                if i.isdigit():
                    c+=1
                else:
                    c=0
            if c==4:
                break
    if c==4: 
        return 1
    else:
        return 0
def onlydigit(x):
    p=len(x)
    if p!=0:
        for i in x:
            if i.isdigit():
                p-=1
        
        if p==0:
            return 1
        else:
            return 0
    else:
        return 0
```

```python
soup = BeautifulSoup(html,'html5lib')
text = soup.get_text(strip=True)
#print(text)
#getting all the text and splitting it by '\n' the spaces and then '.'
tokens = [t for t in text.split('\n')]
finaltokens_v1=[]
finaltokens=[] #list of all tokens after eliminating the unrequired characters
for i in range(len(tokens)):
    tokens[i] = tokens[i].replace('\xa0',' ')
    finaltokens_v1 += list(tokens[i].split(' '))
finaltokens_v1.append('./uln/db/24')
finaltokens_v1.append('12.03.2018')
finaltokens_v1.append('./ul.n/.db/24')
finaltokens_v1.append('2.5.2014')
for i in range(len(finaltokens_v1)):
    #print(list(finaltokens_v1[i].split('.')))
    if checkone(finaltokens_v1[i])==1:
        finaltokens.append(finaltokens_v1[i])
    else:
        finaltokens +=  list(finaltokens_v1[i].split('.'))
#additions for testing
finaltokens.append('December')
finaltokens.append('1968')
finaltokens.append('January')
finaltokens.append('someday')
finaltokens.append('Sunday')
finaltokens.append('someday')
finaltokens.append('Tuesday')
#print(finaltokens)
```

## Date - Month - Year

```python
def date_month_year():
    ##list of the months
    months = ['January','February','March','April','May','June','July','August','September','October','November','December']
    months1 = ['Jan', 'Feb','Mar','Apr','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
    datesl = []
    for i in range(len(finaltokens)):
        for j in months:
            if j in finaltokens[i]:
                #print(i,j)
                datesl.append(i)
    datefinal = [] #length 3 dates (12 dec,1998 || dec 12, 1998)
    datefinal1 = [] #length 2 dates (dec 12 || 12 dec)
    monthYear = []
    onlyMonth = []
    onlyYear = []
    for i in datesl:
        if check1(finaltokens[i-1])==1:
            if check2(finaltokens[i+1])==1:
                datefinal.append(i)
                #print(finaltokens[i-1],finaltokens[i],finaltokens[i+1])
            else:
                datefinal1.append(i)
                #print(finaltokens[i-1],finaltokens[i])
        if len(finaltokens)>i+2:
            if check1(finaltokens[i+1])==1:
                if check2(finaltokens[i+2])==1:
                    datefinal.append(i+1)
                    #print(finaltokens[i],finaltokens[i+1],finaltokens[i+2])
                else:
                    datefinal1.append(i+1)
                    #print(finaltokens[i],finaltokens[i+1])
        if check1(finaltokens[i-1])!=1 and check1(finaltokens[i+1])!=1:
            if check2(finaltokens[i+1])!=1:
                #print(finaltokens[i],"----")
                onlyMonth.append(i)
            else:
                #print('in else')
                monthYear.append(i)
    for i in range(len(finaltokens)):
        if check3(finaltokens[i])==1:
            if finaltokens[i-1] not in months:
                onlyYear.append(i)
                
    datefinal = set(datefinal)
    datefinal = list(datefinal)
    datefinal.sort()
    datefinal1 = set(datefinal1)
    datefinal1 = list(datefinal1)
    datefinal1.sort()
    #print(datefinal)
    #print(datefinal1)
    #print(len(datefinal))
    #print(dates)
    #print(onlyMonth)
    #print(onlyYear)
    p=datefinal+datefinal1+onlyMonth+onlyYear
    p.sort()
    q=[]
    a=[]
    b=[]
    c=[]
    d=[]
    x = 0
    #tagging the dates
    for i in p:
        if i in datefinal:
            #print(finaltokens[i+x],'===')
            finaltokens.insert(i+x-1,'<date>')
            x+=1
            finaltokens.insert(i+2+x,'</date>')
            x+=1
            a.append(i+x)
            q.append(i+x)
        elif i in datefinal1:
            #print(finaltokens[i+x],'===>')
            finaltokens.insert(i+x-1,'<date>')
            x+=1
            finaltokens.insert(i+x+1,'</date>')
            x+=1
            b.append(i+x-1)
            q.append(i+x-1)
        elif i in onlyMonth:
            #print(finaltokens[i])
            finaltokens.insert(i+x,'<Month>')
            x+=1
            finaltokens.insert(i+1+x,'</Month>')
            x+=1
            c.append(i+x-1)
            q.append(i+x-1)
        elif i in onlyYear:
            #print(finaltokens[i])
            finaltokens.insert(i+x,'<Year>')
            x+=1
            finaltokens.insert(i+1+x,'</Year>')
            x+=1
            d.append(i+x-1)
            q.append(i+x-1)
        
    x = 0
    #for i in q:
        #print(finaltokens[i])
    #    if i in a:
    #        print(finaltokens[i+x-3],finaltokens[i+x-2],finaltokens[i+x-1],finaltokens[i+x],finaltokens[i+x+1])
    #    else:
    #        print(finaltokens[i+x-2],finaltokens[i+x-1],finaltokens[i+x],finaltokens[i+x+1])

    dates = 0
    count = 0
    dindex=[]
    #finding dates of the form 12/12/1998 or 12-12-1998 or 12.12.1998
    for i in range(len(finaltokens)):
        count = 0
        if len(finaltokens[i])>=6 and (finaltokens[i].count('/')==2 or finaltokens[i].count('-')==2 or finaltokens[i].count('.')==2):
            #print('in if',i)
            for j in finaltokens[i]:
                if j.isdigit():
                    count+=1
            #print(count,'==>')
            if count>=6:
                #print(finaltokens[i])
                dates += 1
                dindex.append(i)
    #print(dindex)
    #tagging the dates
    y=0
    for i in dindex:
        #print(finaltokens[i+y])
        finaltokens.insert(i+y,'<date>')
        y+=1
        finaltokens.insert(i+y+1,'</date>')
        y+=1
    y=0
    for i in dindex:
        #print(finaltokens[i+y])
        #print(finaltokens[i+y],finaltokens[i+y+1],finaltokens[i+y+2])
        y+=2
    #print(finaltokens[1706])

    #finalhtml = ' '.join(list(finaltokens))
    #print(finalhtml)
```
