```python
from bs4 import BeautifulSoup
import urllib.request
#website to get data from
response = urllib.request.urlopen('https://physics.info/displacement/problems.shtml')
html = response.read()
#print(html)
```

```python
#12.2.14 type dates
def checkone(x):
    count = 0
    c1 =0
    if x.count('.')==2:
        for i in x:
            if i.isdigit():
                count+=1
        if(count>=6):
            for i in range(len(x)):
                if x[i]=='.' and x[i+1].isdigit():
                    if (i-1>=0):
                        if x[i-1].isdigit():
                           c1 +=1
            if c1>0:
                return 1
            else:
                return 0
        else:
            return 0
    else:
        return 0
#day and year check
def check1(x):
    p = 0
    for i in x:
        if i.isdigit():
            p+=1
    if p==2 or p==1:
        return 1
    else:
        return 0
def check2(x):
    p = 0
    for i in x:
        if i.isdigit():
            p+=1
    if p==4:
        return 1
    else:
        return 0
```

```python
soup = BeautifulSoup(html,'html5lib')
text = soup.get_text(strip=True)
#print(text)
#getting all the text and splitting it by '\n' the spaces and then '.'
tokens = [t for t in text.split('\n')]
finaltokens_v1=[]
finaltokens=[] #list of all tokens after eliminating the unrequired characters
for i in range(len(tokens)):
    tokens[i] = tokens[i].replace('\xa0',' ')
    finaltokens_v1 += list(tokens[i].split(' '))
finaltokens_v1.append('./uln/db/24')
finaltokens_v1.append('12.03.2018')
finaltokens_v1.append('./ul.n/.db/24')
finaltokens_v1.append('2.5.2014')
for i in range(len(finaltokens_v1)):
    #print(list(finaltokens_v1[i].split('.')))
    if checkone(finaltokens_v1[i])==1:
        finaltokens.append(finaltokens_v1[i])
    else:
        finaltokens +=  list(finaltokens_v1[i].split('.'))
#additions for testing
finaltokens.append('December')
finaltokens.append('1968')
finaltokens.append('January')
finaltokens.append('someday')
print(finaltokens)
```
